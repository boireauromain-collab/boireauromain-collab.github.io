# Let me merge the 4 years of data to have a big enough dataset

```python
#step 1

import pandas as pd
years = [2022, 2023, 2024, 2025]
qual = pd.concat((pd.read_csv(f"{y}.csv").assign(season=y) for y in years), ignore_index=True)
qual.to_csv("f1_qualifying_2022_2025.csv", index=False)
```

# Average of laps done in qualifying across all seasons

```python
# step 2
import pandas as pd

df = pd.read_csv("f1_qualifying_2022_2025.csv")
avg_laps = df["Laps"].mean()
print(avg_laps)
```
