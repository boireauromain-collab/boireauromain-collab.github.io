Let's visualize the realtionship or lack thereof of two datasets.

We will be looking at two FAO dataset. One on the number of undernourished pople in south Africa; the other on the average supply of protein of animal origin. 

We will be trying to see the trends of each, wether they are similar, what could be expected, etc..

```python
import pandas as pd
import plotly.express as px
```

Let's first take a look at each dataset and reshape them. 
1) Number of people undernourished
2) Average supply of protein of animal origin

```python
undernourished = pd.read_csv("FAO_AS_4473.csv")

undernourished.head()
```

Lots of columns, too much noise. Let's see which column do we need.

```python
undernourished.columns
```

only those ["REF_AREA", "REF_AREA_LABEL","INDICATOR_LABEL", "TIME_PERIOD", "OBS_VALUE"] will suffice. Let's cut the rest out

```python
cols = ["REF_AREA", "REF_AREA_LABEL","INDICATOR_LABEL", "TIME_PERIOD", "OBS_VALUE"]

undernourished_by_country_year = undernourished[cols]

undernourished_by_country_year
```

Now let's only keep the country we are interested in: South Africa

```python
undernourished_south_africa = undernourished_by_country_year[undernourished_by_country_year["REF_AREA_LABEL"] == "South Africa"]
undernourished_south_africa
```

let's take a look at the evolution of undernourished people in South Africa

```python
fig = px.line(
    undernourished_south_africa,
    x="TIME_PERIOD",
    y="OBS_VALUE",
    title="Number of Undernourished People in South Africa Over Time",
    labels={"TIME_PERIOD": "Year", "OBS_VALUE": "Number of Undernourished People"},
)
fig.show()
```

Let's repeapt the same steps for the other dataset: the average supply of protein of animal origin

```python
protein= pd.read_csv("FAO_FS_21014.csv")

protein.head()
```

```python
protein.columns
```

```python
cols = ["REF_AREA", "REF_AREA_LABEL","INDICATOR_LABEL", "TIME_PERIOD", "OBS_VALUE"]

protein_by_country_year = protein[cols]
protein_by_country_year
```

```python
protein_South_africa = protein_by_country_year[protein_by_country_year["REF_AREA_LABEL"] == "South Africa"]
protein_South_africa
```

```python
fig = px.line(
    protein_South_africa,
    x="TIME_PERIOD",
    y="OBS_VALUE",
    title="Average Supply of Protein of Animal Origin in South Africa Over Time",
    labels={"TIME_PERIOD": "Year", "OBS_VALUE": "Average Supply of Protein of Animal Origin (g/capita/day)"},
)
fig.show()
```

Let us now plot the two reworked dataset together

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

country_name = "South Africa"

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(
        x=undernourished_south_africa["TIME_PERIOD"],
        y=undernourished_south_africa["OBS_VALUE"],
        name="Number of Undernourished People",
    ),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(
        x=protein_South_africa["TIME_PERIOD"],
        y=protein_South_africa["OBS_VALUE"],
        name="Average Supply of Protein of Animal Origin",
    ),
    secondary_y=True,
)

fig.update_layout(title_text=f"Number of Undernourished People vs. Average Supply of Protein of Animal Origin, {country_name}")
fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Number of Undernourished People", secondary_y=False)
fig.update_yaxes(title_text="Average Supply of Animal Protein (g/capita/day)", secondary_y=True)

fig.show()
```

One could have expected that the increase trend in average supply of animal protein in South Africa could have helped decrease the number of people undernourished. Surprisingly, both have increased in the same direction.

The two line shows a similar slope and even share a dip at the same moment in time. This might be an indicator that there is a correlation between the two observations or that both are affected by a common event but in any case we obviously cannot draw any causal conclusions as we missed a lot of control variables but there is a visiual similitude.
