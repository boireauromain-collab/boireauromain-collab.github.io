# Now the hardway: average, median, mode

```python
# step 4
import csv
from statistics import mean, median, mode

delays = []
skipped = 0

with open("2017.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        s = (row.get("ArrDelay") or "").strip()
        if s == "" or s.lower() in {"na", "n/a", "none"}:
            skipped += 1
            continue
        delays.append(float(s))

print(f"Kept {len(delays)} rows, skipped {skipped} empty/non-numeric rows.")
print("Average delay:", mean(delays))
print("Median delay:", median(delays))
print("Mode delay(s):", mode(delays))
```

# Visualization

```python
import csv

# 1) Load delays (floats), skip empties
delays = []
with open("2017.csv", newline="", encoding="utf-8-sig") as f:
    for r in csv.DictReader(f):
        s = (r.get("ArrDelay") or "").strip()
        if s: delays.append(float(s))


bins = list(range(-60, 181, 30))
counts = [0]*(len(bins)-1)
for x in delays:
    i = int((x - bins[0]) // 30)
    i = max(0, min(len(counts)-1, i))
    counts[i] += 1


print("Arrival Delay Histogram (minutes)\n")
m = max(counts) or 1
for i, c in enumerate(counts):
    left, right = bins[i], bins[i+1]
    bar = "#" * (c * 40 // m)      
    print(f"[{left:>4}..{right-1:>4}] | {bar} {c}")
print("\nFlights per bin  |  Delay (min) →")
```
